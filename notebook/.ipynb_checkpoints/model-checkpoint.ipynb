```python
data_list = []
for t in text:
    t = re.sub(r'[!@#$(),\n"%^*?\:;~`0-9]', ' ', t)
    t = re.sub(r'[[]]', ' ', t)
    t = t.lower()
    data_list.append(t)
```

```python
# preprocessed_df = text_preprocess(df,'Text')
# preprocessed_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text, language, test_size = 0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
count_vectorizer.fit(X_train)

x_train = count_vectorizer.transform(X_train).toarray()
x_test  = count_vectorizer.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

ac = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
```

```python
print("Accuracy is :",ac)
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([('vectorizer', count_vectorizer), ('multinomialNB', model)])
pipe.fit(X_train, y_train)
```

```python
y_pred2 = pipe.predict(X_test)
ac2 = accuracy_score(y_test, y_pred2)
print("Accuracy is :",ac2)
```

```python
with open('../src/model/trained_pipeline-0.1.0.pkl','wb') as f:
    pickle.dump(pipe, f)
```

```python
text = "Hello, how are you?"
text = "Ciao, come stai?"
language = pipe.predict([text])
print(label_encoder.classes_[language[0]], language)
```
